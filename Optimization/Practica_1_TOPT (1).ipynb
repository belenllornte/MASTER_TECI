a.) Si se quiere acotar la solución en un intervalo de magnitud $0.01$, ¿cuál es el número de iteraciones necesarias con el algoritmo de la sección áurea? Implementa dicho algoritmo en una función para resolver el problema.


```python
epsilon = 0.01 #tolerancia propuesta en el enunciado
```

```python
def seccion_aurea(x_min, x_max, eps):
    phi = (1 + math.sqrt(5)) / 2  # número áureo ~ 1.618

    # Inicializar puntos (la distancia entre ellos y los extremos del intervalo esté en proporción áurea):
    a = x_max -  ((x_max - x_min)/ phi)
    b = x_min + ((x_max - x_min)/ phi)

    iter = 0 # parámetro que se irá actualizando, contando así el número de iteracciones necesarias.


    while abs(x_max - x_min) > eps:
        if f(a) < f(b):
            x_max = b
            b = a
            a = x_max - ((x_max - x_min) / phi)
        else:
            x_min = a
            a = b
            b = x_min + ((x_max - x_min) / phi)


        # Hacemos la siguiente iteración
        iter += 1


    return x_min, x_max,(x_max + x_min) / 2, iter
```

```python
sol_aurea = seccion_aurea(a, b, epsilon)
print(f"Sección áurea: Mínimo aproximado entre x_L={sol_aurea[0]} y x_U={sol_aurea[1]} es en: x = {sol_aurea[2]} con {sol_aurea[3]} iteraciones.")
```

b.) Con la misma amplitud $0.01$ del intervalo final, ¿cuál sería el número de iteraciones necesarias con el algoritmo de Fibonacci? Implementa dicho algoritmo en una función para resolver el problema.


```python
# Definimos una función que devuelva el n-ésimo número de Fibonacci, con condiciones F(0) = 0, F(1) = 1 y F(2) = 2
def Fibonacci(n):
    if n < 0:
      print("Introduzca un número no negativo: ")
    elif n == 0:
      return 0
    elif n == 1:
      return 1
    elif n == 2:
      return 2
    else:
      return Fibonacci(n-1) + Fibonacci(n-2)
```

```python
def metodo_Fibonacci(x_min,x_max,eps):
  '''
  - x_min: Límite inferior del intervalo
  - x_max: Límite superior del intervali
  - eps: Tolerancia
  '''

  I_0 = x_max - x_min # Longitud inicial del intervalo
  F_0 = 0
  F_1 = 1
  ind_fibonacci = 0
  F_I = Fibonacci(ind_fibonacci)

  # Encuentra el número mínimo en la secuencia de Fibonacci que permita dividir
  # el intervalo inicial con la precisión eps (Fibonacci(k) > I_0 / eps).
  while (F_I < I_0/eps):
    ind_fibonacci += 1
    F_I = Fibonacci(ind_fibonacci)

  i = 0
  # Si el índice es mayor que 2, empieza el proceso.
  if ind_fibonacci > 2:
    # Calcular los puntos de prueba iniciales x_a y x_b dentro del intervalo [x_min, x_max].
    x_a = x_min + Fibonacci(ind_fibonacci-2)/Fibonacci(ind_fibonacci)*(x_max-x_min)
    f_a = f(x_a)
    x_b = x_max - Fibonacci(ind_fibonacci-2)/Fibonacci(ind_fibonacci)*(x_max-x_min)
    f_b = f(x_b)
    # Empieza el proceso de reducción de intervalo hasta que la precisión deseada se alcance.
    while(i <= (ind_fibonacci-3)):
      if (f_a < f_b):
        # Si f(x_a) < f(x_b), entonces el mínimo está entre x_min y x_b.
        x_max = x_b
        x_b = x_a
        f_b = f_a
        x_a = x_min + Fibonacci(ind_fibonacci - i - 2)/Fibonacci(ind_fibonacci - i)*(x_max-x_min)
        f_a = f(x_a)
      elif (f_a > f_b):
        # Si f(x_a) > f(x_b), entonces el mínimo está entre x_a y x_max.
        x_max = x_a
        x_a = x_b
        f_a = f_b
        x_b = x_max - Fibonacci(ind_fibonacci - i - 2)/Fibonacci(ind_fibonacci - i)*(x_max-x_min)
        f_b = f(x_b)
      else:
        # Si f(x_a) == f(x_b), ajusta ambos límites.
        x_min = x_a
        x_max = x_b
        i = i + 2 # Incrementa en 2 ya que hemos ajustado ambos puntos.
        if (i < (ind_fibonacci - 2)):
          # Recalcula x_a y x_b si quedan iteraciones por hacer.
          x_a = x_min + Fibonacci(ind_fibonacci - i - 2)/Fibonacci(ind_fibonacci - i)*(x_max-x_min)
          f_a = f(x_a)
          x_b = x_max - Fibonacci(ind_fibonacci - i - 2)/Fibonacci(ind_fibonacci - i)*(x_max-x_min)
          f_b = f(x_b)
      i += 1
    if (i == ind_fibonacci - 2):
      if (f_a < f_b):
        # Ajusta el límite superior si f_a es menor.
        x_max = x_b
        i+=1
      elif (f_a > f_b):
        # Ajusta el límite inferior si f_b es menor.
        x_min = x_a
        i+=1
      else:
        # Ajusta ambos límites si son iguales.
        x_min = x_a
        x_max = x_b
        i+=2

  # Al llegar a la última iteración, establece el intervalo final en el centro.
  if (i == ind_fibonacci - 1):
    x_a = (x_min + x_max)/2
    x_b = (x_min + x_max)/2
    # Ajusta el intervalo según el valor de la función en los puntos medios.
    if (f(x_a) <= f(x_b)):
      x_max = x_b
    else:
      x_min = x_a
  # Devuelve los límites del intervalo de búsqueda y el punto medio.
  return x_min,x_max,(x_min + x_max)/2, i

sol_Fibonacci = metodo_Fibonacci(a,b,epsilon)
```

b.) Desarrollar un par de iteraciones con el algoritmo de Steepest descent.


```python
def steepest_descent(f, f_grad, x0, lamda=0.1, max_iters=2):
    '''
    - f: Función a minimizar.
    - f_grad: Gradiente de la función a minimizar.
    - x0: Punto de partida.
    - lamda: Valor inicial de λ.
    - max_iters: Número de iteraciones máximas (en este enunciado nos piden 2, pero se podría modificar).
    '''
    x = np.array(x0)
     # Bucle principal para las iteraciones del método.
    for i in range(max_iters):
        grad_x = f_grad(x)
        # Calculamos al dirección de máximo descenso
        max_des = -grad_x / np.linalg.norm(grad_x)

        # Definimos una función f_lambda que calcula f(x + λ * dirección_descenso) para un valor de λ.
        # Esto se usará para optimizar el valor de λ en cada paso.
        def f_lambda(lamda):
            new_x = x + lamda * max_des
            return f(new_x[0], new_x[1])

        res = minimize(f_lambda, lamda) # Calculamos el valor óptimo de λ minimizando f_lambda
        lambda_optimo = res.x # Almacenamos el valor óptimo de λ

        x = x + lambda_optimo * max_des # Actualizamos el punto x

        print(f"Iteración {i+1}: x = {x}, dirección de máximo descenso = {max_des}, λ óptimo = {lambda_optimo}. \n")

    return x
```

c.) Desarrollar un par de iteraciones con el algoritmo de Newton.


```python
# De manera muy parecida a lo anterior con el gradiente hacemos con la hessiana:

def hessiana(f, lista_de_variables):
    return [[diff(g, var) for var in lista_de_variables] for g in gradiente(f,lista_de_variables)]

hessiana_f3 = hessiana(f3_simbolica, [x_sym, y_sym])
print(hessiana_f3)


def sympy_to_numpy_matriz(expression, symbols):
    """
    Convierte una matriz de SymPy en una matriz de funciones NumPy.
    Actúa de manera recursiva.
    """
    # Si 'expression' es una matriz (lista o tuple), aplicamos la función recursivamente a cada elemento.
    if isinstance(expression, (list, tuple)):
        return [sympy_to_numpy_matriz(expr, symbols) for expr in expression]  # Llamada recursiva para cada elemento
    else:
    # Si 'expression' no es una matriz, significa que es una expresión individual de SymPy.
        return lambdify(symbols, expression, 'numpy')  # Convertir la expresión a una función de NumPy

# Convertir la hessiana a una función numérica
hess_f_numeric = [sympy_to_numpy_matriz(g, [x_sym, y_sym]) for g in hessiana_f3]


def f_hess(x):
    # Evaluar cada función en la matriz de la hessiana en el punto x
    return np.array([[h(x[0], x[1]) for h in row] for row in hess_f_numeric])
```

```python
# Mirando los apuntes del tema correspondiente observamos que el algoritmo de Newton para encontrar un mínimo de una función f(x) es:
# x^(k+1) = x^k - H(x^k)^-1 * grad(f(x^k)), donde x^k es el punto actual donde se evalúa el método.
# H(x^k)^-1 es la inversa de la hessiana de la función en dicho punto
# grad(f(x^k)) gradiente de la función en el punto a evaluar
# x^k es el punto actual, el cual se está evaluando
# x(k+1) es el siguiente punto de la iteración del método
# de este modo en el algoritmo siguiente denominamos como es de constumbre delta a los incrementos, así x^(k+1) - x^k será deta_x (el incremento de los x, es decir, el incremento de una iteracción a otra)
# desspejando se nos queda: x^(k+1) - x^k = - H(x^k)^-1 * grad(f(x^k)) ---> delta_x = - H(x^k)^-1 * grad(f(x^k))
# para no hacer uso de la inversa de la hessiana nos quedamos con:  H(x^k) * delta_x = -grad(f(x^k))
# y para resolver este sistema hacemos uso de la funcion solve de la libreria numpy. De esta manera obtenemos el vector delta_x, pero como conocemos x^k, entonces tendremos x_(k+1)


def newton(f,f_grad, f_hess, x0, max_iters=2):

    '''
    - f: Función a minimizar.
    - f_grad: Gradiente de la función a minimizar.
    - f_hess: Hessiana de la función a minimizar.
    - x0: Punto de partida.
    - max_iters: Número de iteraciones máximas (en este enunciado nos piden 2, pero se podría modificar).
    '''

    x=np.array(x0).astype(float)

    for iter in range(max_iters):
        # gradiente y hessiana en el punto a evaluar
        gradi = f_grad(x)
        hessi = f_hess(x)

        # Resolvemos el sistema  H(x^k) * delta_x = -grad(f(x^k)) para obtener delta_x (ya explicado previamente cómo llegamos a este resultado) --pensar como Ax=b -> x=solve(A,b)
        delta_x = np.linalg.solve(hessi, -gradi)

        # Como sabemos delta_x y x^k, podemos calcular x^(k+1) de la siguiente manera:
        x1 = x + delta_x

        x=x1

        print(f"Iteración {iter+1}: x = {x}, gradiente = {gradi}, hessiana={hessi},delta_x = {delta_x}\n") # en este caso concreto el delta_x y el x coinciden porque el punto inicial es el (0,0)

    return x
```
